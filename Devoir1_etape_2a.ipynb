### Code préliminaire d'accès aux données et d'import

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading data from the csv file
iris_data = pd.read_csv('data/iris.csv')

variables_iris = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

X = iris_data[variables_iris]
Y = iris_data[['species']]

X_setosa = X.loc[Y['species'] == 'setosa']
X_versicolor = X.loc[Y['species'] == 'versicolor']
X_virginica = X.loc[Y['species'] == 'virginica']
```

## Section 1 : Affichage de chaque classe individuellement selon chaque variable <a name="section1"></a>

```python
# Objet pour afficher des histogrammes pour les classes individuelles de la section 1.
class individual_hist:
    
    def __init__(self, X_Data, class_name):
        self.X_Data = X_Data
        self.class_name = class_name
    
    
    def hist(self, variable1):
        plt.hist(self.X_Data[variable1], ec='black')
        
        plt.title(f"Classe individuelle: {self.class_name}")
        plt.xlabel(variable1)
        
        plt.show()
```

### 1.1 setosa <a name="section1.1"></a>

```python
ind_setosa = individual_hist(X_Data=X_setosa, class_name='setosa')

for variable in variables_iris:
    ind_setosa.hist(variable)
```

### 1.2 versicolor <a name="section1.2"></a>

```python
ind_versicolor = individual_hist(X_Data=X_versicolor, class_name='versicolor')

for variable in variables_iris:
    ind_versicolor.hist(variable)
        
```

### 1.3 virginica <a name="section1.3"></a>

```python
ind_virginica = individual_hist(X_Data=X_virginica, class_name='virginica')

for variable in variables_iris:
    ind_virginica.hist(variable)
```

## Section 2 : Affichage de chaque duo de classes selon chaque variable <a name="section2"></a>

```python
# Objet pour afficher des histogrammes pour les classes en duo de la section 2.
class duo_hist:
    
    def __init__(self, X_Data_class1, X_Data_class2, class_name1, class_name2):
        self.X_Data_class1 = X_Data_class1
        self.class_name1 = class_name1
        
        self.X_Data_class2 = X_Data_class2
        self.class_name2 = class_name2
    
    
    def hist(self, variable1):
        plt.hist([self.X_Data_class1[variable1], self.X_Data_class2[variable1]], 
                 label=[self.class_name1,self.class_name2], ec='black', stacked=True)
        
        plt.title(f"Classes: {self.class_name1} vs {self.class_name2}")
        plt.xlabel(variable1)
        
        plt.legend()
        plt.show()
```

### 2.1 : setosa vs versicolor <a name="section2.1"></a>

```python
duo_set_vers_hist = duo_hist(X_Data_class1=X_setosa,
                               X_Data_class2=X_versicolor,
                               class_name1='setosa',
                               class_name2='versicolor')

for variable in variables_iris:
    duo_set_vers_hist.hist(variable)
```

### 2.2 : setosa vs virginica <a name="section2.2"></a>

```python
duo_set_virg_hist = duo_hist(X_Data_class1=X_setosa,
                               X_Data_class2=X_virginica,
                               class_name1='setosa',
                               class_name2='virginica')

for variable in variables_iris:
    duo_set_virg_hist.hist(variable)
```

### 2.3 : versicolor vs virginica <a name="section2.3"></a>

```python
duo_vers_virg_hist = duo_hist(X_Data_class1=X_versicolor,
                               X_Data_class2=X_virginica,
                               class_name1='versicolor',
                               class_name2='virginica')

for variable in variables_iris:
    duo_vers_virg_hist.hist(variable)
```

## Section 3 : Affichage des trois classes selon chaque variable <a name="section3"></a>

```python
for variable in variables_iris:    
        plt.hist([X_setosa[variable], X_versicolor[variable], X_virginica[variable]], 
                 label=['setosa', 'versicolor', 'virginica'], 
                 color=['b', 'g', 'k'],
                 ec='black', 
                 stacked=True)
        
        plt.title("Visualisation de toutes les classes")
        plt.xlabel(variable)

        plt.legend()
        plt.show()
```
