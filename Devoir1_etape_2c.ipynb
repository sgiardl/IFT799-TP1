### Code préliminaire d'accès aux données et d'import

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading data from the csv file
iris_data = pd.read_csv('data/iris.csv')

variables_iris = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

X = iris_data[variables_iris]
Y = iris_data[['species']]

X_setosa = X.loc[Y['species'] == 'setosa']
X_versicolor = X.loc[Y['species'] == 'versicolor']
X_virginica = X.loc[Y['species'] == 'virginica']
```

## Section 1 : Affichage de chaque classe individuellement selon chaque duo de variables <a name="section1"></a>

```python
# Objet pour afficher des scatterplots pour les classes individuelles de la section 1.
class individual_scatter:
    
    def __init__(self, X_Data, class_name):
        self.X_Data = X_Data
        self.class_name = class_name
    
    
    def scatter(self, variable1, variable2):
        plt.scatter(self.X_Data[variable1], self.X_Data[variable2])
        
        plt.title(f"Classe individuelle: {self.class_name}")
        plt.xlabel(variable1)
        plt.ylabel(variable2)
        
        plt.show()
```

### 1.1 setosa <a name="section1.1"></a>

```python
ind_setosa = individual_scatter(X_Data=X_setosa, class_name='setosa')

for i in range(len(variables_iris)):
    for j in range(i+1, len(variables_iris)):
        ind_setosa.scatter(variables_iris[i], variables_iris[j])
```

### 1.2 versicolor <a name="section1.2"></a>

```python
ind_versicolor = individual_scatter(X_Data=X_versicolor, class_name='versicolor')

for i in range(len(variables_iris)):
    for j in range(i+1, len(variables_iris)):
        ind_versicolor.scatter(variables_iris[i], variables_iris[j])
```

### 1.3 virginica <a name="section1.3"></a>

```python
ind_virginica = individual_scatter(X_Data=X_virginica, class_name='virginica')

for i in range(len(variables_iris)):
    for j in range(i+1, len(variables_iris)):
        ind_virginica.scatter(variables_iris[i], variables_iris[j])
```

## Section 2 : Affichage de chaque duo de classes selon chaque duo de variables <a name="section2"></a>

```python
# Objet pour afficher des scatterplots pour les classes en duo de la section 2.
class duo_scatter:
    
    def __init__(self, X_Data_class1, X_Data_class2, class_name1, class_name2):
        self.X_Data_class1 = X_Data_class1
        self.class_name1 = class_name1
        
        self.X_Data_class2 = X_Data_class2
        self.class_name2 = class_name2
    
    
    def scatter(self, variable1, variable2):
        plt.scatter(self.X_Data_class1[variable1], self.X_Data_class1[variable2], label=self.class_name1)
        plt.scatter(self.X_Data_class2[variable1], self.X_Data_class2[variable2], label=self.class_name2)
        
        plt.title(f"Classes: {self.class_name1} vs {self.class_name2}")
        plt.xlabel(variable1)
        plt.ylabel(variable2)
        
        plt.legend()
        plt.show()
```

### 2.1 : setosa vs versicolor <a name="section2.1"></a>

```python
duo_set_vers_scatter = duo_scatter(X_Data_class1=X_setosa,
                               X_Data_class2=X_versicolor,
                               class_name1='setosa',
                               class_name2='versicolor')

for i in range(len(variables_iris)):
    for j in range(i+1, len(variables_iris)):
        duo_set_vers_scatter.scatter(variables_iris[i], variables_iris[j])
```

### 2.2 : setosa vs virginica <a name="section2.2"></a>

```python
duo_set_virg_scatter = duo_scatter(X_Data_class1=X_setosa,
                               X_Data_class2=X_virginica,
                               class_name1='setosa',
                               class_name2='virginica')

for i in range(len(variables_iris)):
    for j in range(i+1, len(variables_iris)):
        duo_set_virg_scatter.scatter(variables_iris[i], variables_iris[j])
```

### 2.3 : versicolor vs virginica <a name="section2.3"></a>

```python
duo_vers_virg_scatter = duo_scatter(X_Data_class1=X_versicolor,
                               X_Data_class2=X_virginica,
                               class_name1='versicolor',
                               class_name2='virginica')

for i in range(len(variables_iris)):
    for j in range(i+1, len(variables_iris)):
        duo_vers_virg_scatter.scatter(variables_iris[i], variables_iris[j])
```

## Section 3 : Affichage des trois classes selon chaque duo de variables <a name="section3"></a>

```python
for i in range(len(variables_iris)):
    for j in range(i+1, len(variables_iris)):    
        variable1 = variables_iris[i]
        variable2 = variables_iris[j]
            
        plt.scatter(X_setosa[variable1], X_setosa[variable2], label='setosa', color='b')
        plt.scatter(X_versicolor[variable1], X_versicolor[variable2], label='versicolor', color='g')
        plt.scatter(X_virginica[variable1], X_virginica[variable2], label='virginica', color='k')

        plt.title("Visualisation de toutes les classes")
        plt.xlabel(variable1)
        plt.ylabel(variable2)

        plt.legend()
        plt.show()
```
