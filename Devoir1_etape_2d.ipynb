### Code préliminaire d'accès aux données et d'import

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA  # https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html
```

```python
# Loading data from the csv file
iris_data = pd.read_csv('data/iris.csv')

variables_iris = np.array(['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])

X = iris_data[variables_iris]
Y = iris_data[['species']]
```

```python
# On construit un nouvel ensemble de données avec une variable transformée (selon les 4 variables de base)

pca = PCA(n_components=2)
X_transform = pd.DataFrame(pca.fit_transform(X))
```

## Section 1 : Affichage de chaque classe individuellement selon les variables transformees <a name="section1"></a>

### 1.1 setosa <a name="section1.1"></a>

```python
plt.scatter(X_transform.loc[Y['species'] == 'setosa'][0], X_transform.loc[Y['species'] == 'setosa'][1], color='b')

plt.title(f"Classe individuelle: setosa")
plt.xlabel("variable1")
plt.ylabel("variable2")

plt.show()
```

### 1.2 versicolor <a name="section1.2"></a>

```python
plt.scatter(X_transform.loc[Y['species'] == 'versicolor'][0], X_transform.loc[Y['species'] == 'versicolor'][1], color='g')

plt.title(f"Classe individuelle: versicolor")
plt.xlabel("variable1")
plt.ylabel("variable2")

plt.show()
```

### 1.3 virginica <a name="section1.3"></a>

```python
plt.scatter(X_transform.loc[Y['species'] == 'virginica'][0], X_transform.loc[Y['species'] == 'virginica'][1], color='k')

plt.title(f"Classe individuelle: virginica")
plt.xlabel("variable1")
plt.ylabel("variable2")

plt.show()
```

## Section 2 : Affichage de chaque duo de classes selon les variables transformees <a name="section2"></a>

### 2.1 : setosa vs versicolor <a name="section2.1"></a>

```python
plt.scatter(X_transform.loc[Y['species'] == 'setosa'][0], X_transform.loc[Y['species'] == 'setosa'][1], label='setosa', color='b')
plt.scatter(X_transform.loc[Y['species'] == 'versicolor'][0], X_transform.loc[Y['species'] == 'versicolor'][1], label='versicolor', color='g')

plt.title(f"Classes: setosa vs versicolor")
plt.xlabel("variable1")
plt.ylabel("variable2")

plt.legend()
plt.show()
```

### 2.2 : setosa vs virginica <a name="section2.2"></a>

```python
plt.scatter(X_transform.loc[Y['species'] == 'setosa'][0], X_transform.loc[Y['species'] == 'setosa'][1], label='setosa', color='b')
plt.scatter(X_transform.loc[Y['species'] == 'virginica'][0], X_transform.loc[Y['species'] == 'virginica'][1], label='virginica', color='k')

plt.title(f"Classes: setosa vs virginica")
plt.xlabel("variable1")
plt.ylabel("variable2")

plt.legend()
plt.show()
```

### 2.3 : versicolor vs virginica <a name="section2.3"></a>

```python
plt.scatter(X_transform.loc[Y['species'] == 'versicolor'][0], X_transform.loc[Y['species'] == 'versicolor'][1], label='versicolor', color='g')
plt.scatter(X_transform.loc[Y['species'] == 'virginica'][0], X_transform.loc[Y['species'] == 'virginica'][1], label='virginica', color='k')

plt.title(f"Classes: versicolor vs virginica")
plt.xlabel("variable1")
plt.ylabel("variable2")

plt.legend()
plt.show()
```

## Section 3 : Affichage des trois classes selon les variables transformees <a name="section3"></a>

```python
plt.scatter(X_transform.loc[Y['species'] == 'setosa'][0], X_transform.loc[Y['species'] == 'setosa'][1], label='setosa', color='b')
plt.scatter(X_transform.loc[Y['species'] == 'versicolor'][0], X_transform.loc[Y['species'] == 'versicolor'][1], label='versicolor', color='g')
plt.scatter(X_transform.loc[Y['species'] == 'virginica'][0], X_transform.loc[Y['species'] == 'virginica'][1], label='virginica', color='k')

plt.title(f"Classes: setosa vs versicolor vs virginica")
plt.xlabel("variable1")
plt.ylabel("variable2")

plt.legend()
plt.show()
```

## Section 4 : Affichage des trois classes selon chaque combinaison des variables <a name="section4"></a>

```python
# Ici on tente de voir l'impact d'enlever des variables dans la variable transformée

import itertools
for L in range(3, len(variables_iris)):
    for subset in itertools.combinations(variables_iris, L):
        chosen_variables = list(subset)
        
        # On produit la variable transformée selon variables choisies
        X_transform = pd.DataFrame(pca.fit_transform(X[chosen_variables]))  
        
        # On produit le scatterplot
        plt.scatter(X_transform.loc[Y['species'] == 'setosa'][0], X_transform.loc[Y['species'] == 'setosa'][1], label='setosa', color='b')
        plt.scatter(X_transform.loc[Y['species'] == 'versicolor'][0], X_transform.loc[Y['species'] == 'versicolor'][1], label='versicolor', color='g')
        plt.scatter(X_transform.loc[Y['species'] == 'virginica'][0], X_transform.loc[Y['species'] == 'virginica'][1], label='virginica', color='k')
        
        plt.title(f"Comparaison des trois classes selon les variables {subset}")
        plt.xlabel(f"Variable transformee selon les variables {subset}")
        plt.legend()

        plt.show()
```
